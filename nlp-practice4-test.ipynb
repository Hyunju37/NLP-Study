```python
from transformers import AutoTokenizer
tokenizer1 = AutoTokenizer.from_pretrained("google-bert/bert-base-uncased")
tokenizer2 = AutoTokenizer.from_pretrained("distilbert-base-uncased")
tokenizer3 = AutoTokenizer.from_pretrained("t5-base")
tokenizer4 = AutoTokenizer.from_pretrained("gpt2")
tokenizer5 = AutoTokenizer.from_pretrained('roberta-base')
```

```python
encoding1 = tokenizer1.encode(text=sample_q)
encoding2 = tokenizer2.encode(text=sample_q)
encoding3 = tokenizer3.encode(text=sample_q)
encoding4 = tokenizer4.encode(text=sample_q)
encoding5 = tokenizer5.encode(text=sample_q)
```

```python
print(tokenizer1.convert_ids_to_tokens(encoding1))
print(tokenizer2.convert_ids_to_tokens(encoding2))
print(tokenizer3.convert_ids_to_tokens(encoding3))
print(tokenizer4.convert_ids_to_tokens(encoding4))
print(tokenizer5.convert_ids_to_tokens(encoding5))
```
